```python
import pandas as pd

wk_dir = '/Users/ltamon/SahakyanLab/GenomicContactDynamics/26_PredictingCp'
df = pd.read_csv(wk_dir + '/out_makeTable/min2Mb_chr21_MLtbl.csv')

import re

def choose_features(df, feat_regex, label_col):
    
    features = np.array(df.columns)
    is_chosen = [ bool(re.search(pattern=feat_regex, string=feat)) for feat in features ] 
    
    return( df[ np.append(label_col, features[is_chosen]) ] )
    
import numpy as np

def switch_contact(df):
    
    features = np.array(df.columns)
    is_icol = ['.i_' in feat for feat in features]
    is_jcol = ['.j_' in feat for feat in features]
    
    features_switch = np.array(features)
    features_switch[is_icol] = features[is_jcol]
    features_switch[is_jcol] = features[is_icol]
    
    df_switch = df.set_axis(features_switch, axis=1, inplace=False)
    df_switch = df_switch[features]
    
    return( pd.concat([df, df_switch]) )
```

```python
feat_regex = 'grp.compl.'

df = choose_features(df, feat_regex, 'label')
switch_contact(df)
```
